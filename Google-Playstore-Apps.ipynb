```python
# imports
import pandas as pd
```

```python
# load data
url = 'https://media.githubusercontent.com/media/gauthamp10/Google-Playstore-Dataset/main/Google-Playstore.csv'

df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
# any null values?
df.isnull().sum()
```

```python
# list columns
df.columns
```

```python
# top 50 apps by Rating Count
df.sort_values(by='Rating Count', ascending=False).head(50)
```

```python
# top 50 apps by installs
df.sort_values(by='Installs', ascending=False).head(50)
```

```python
# how many of each threshold of installs
df['Installs'].value_counts()
```

```python
# category
df['Category'].value_counts(normalize=True).head(15)
```

```python
# content rating
df['Content Rating'].value_counts(normalize=True)
```

```python
# free to download
df['Free'].value_counts(normalize=True)
```

```python
# ad supported
df['Ad Supported'].value_counts(normalize=True)
```

```python
# in app purchases
df['In App Purchases'].value_counts(normalize=True)
```

```python
# editors choice
df['Editors Choice'].value_counts(normalize=True)
```
